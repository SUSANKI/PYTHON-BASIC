```python
# Let create some fake dataset and draw the line chart @-@
x=np.linspace(0,30,1500)
y= 0.2*x + np.sin(x)
```

### SCATTER PLOT

```python
# plt.scatter()
```

NOTE: 
    
    Scatter plot is useful in determining groups and classification of data, hence it is applied in Classification and Clustering prblems
    
    For data of high dimensional (>3), then we can use dimensional reduction to turn data into lower dimension and apply the scatter plot

```python
# Lets make a fake dataset for scatter plot
x=np.linspace(1,10,500)
y= 2*x + 2*np.random.randn(500)
```

```python
plt.scatter(x,y,c='darkblue')
plt.title('SCATTER PLOT');
```

```python
# Lets make a new dataset with 3 different clusters @-@, but how ??
x=np.linspace(0,20,1200)
y=np.random.randn(1200)
y[:400] += 3
y[800:] += 5
label=np.zeros(1200)
label[400:800]=1 ; label[800:]=2
```

```python
plt.scatter(x,y,c=label)
```

```python
x=np.random.randn(500,2)
label=np.zeros(500); label[:250]=1
x[:250,] += 3
```

```python
plt.scatter(x[:,0],x[:,1],c=label)
```

### HISTOGRAM

```python
# plt.hist()
```

NOTE:
    
    Histogram is widely used to visualize the probability distribution of the dataset
    
    It can be used for multi-dimensional data, but only 1 and 2 dimension can be visualize => dimensionality reduction

```python
x=np.random.normal(2,3,1000)
```

```python
plt.hist(x,bins=50,color='darkmagenta');
```

```python
# What if there are 2 dimensional dataset, can the hist function generate histogram in 3 dimension @-@
x=np.random.randn(10000,2)
plt.hist(x,bins=25,color=('red','darkgreen'))
plt.show()
```

```python
# AMAZING !!!
# So u can guess about 3 dimensional data @-@, huh??
x= np.random.randn(10000,3)
plt.hist(x,bins=20, color=('red','green','deeppink'))
plt.show()
```

#### EXERCISE OF MATPLOTLIB

```python
x=np.random.uniform(-1,1,1400).reshape(700,2)
```

```python
label=np.zeros(700)
for i in range(700):
    if((x[i,0]<0) & (x[i,1]<0)) | ((x[i,0]>=0) & (x[i,1]>=0)):
        label[i]=1
   
```

```python
plt.scatter(x[:,0],x[:,1],c=label)
```
